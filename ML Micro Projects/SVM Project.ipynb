```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("train.csv")
```

```python
dataset.columns
```

```python
from sklearn import preprocessing
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import svm
```

```python
dataset
```

```python
le = preprocessing.LabelEncoder()
```

```python
dataset["Sex"] = le.fit_transform(dataset["Sex"])
```

```python
dataset["Embarked"] = le.fit_transform(dataset["Embarked"])
```

```python
dataset
```

```python
x1 = dataset.drop(["Name", "PassengerId", "Ticket", "Cabin"],axis=1)
```

```python
x1
```

```python
import numpy as np
x1["Age"] = x1["Age"].astype(int)
x1["Fare"] = x1["Fare"].astype(int)
```

```python
for a in x1.columns:
    y = x1[a]
    x2 = x1.drop([a],axis=1)
    x_train, x_test, y_train, y_test = train_test_split(x2, y, test_size = 0.3, random_state = 0)
    clf = svm.SVC(gamma=0.01, C=100)
    print("",clf.fit(x_train, y_train))
    y_pred = clf.predict(x_test)
    print("\n accuracy_scorec for ",a,":",accuracy_score(y_test, y_pred, normalize=True))
    print("confusion matrix for",a,":\n",confusion_matrix(y_test, y_pred))
```
