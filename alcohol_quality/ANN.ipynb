```python
import pandas as pd
import numpy as np
import keras
from keras.layers import Dense, LeakyReLU, PReLU, ELU, Dropout
from keras.models import Sequential
```

```python
data = pd.read_csv('alcohol-quality-data.csv')
```

```python
dataset = data.copy()
```

```python
y = pd.DataFrame(dataset['Quality_Category'].replace({"Low":0, "High":1}))
x = dataset.drop(columns=['Quality_Category'])
```

```python
x
```

```python
x_new = x.drop(['pH', 'density'], axis=1)
```

```python
model = Sequential()
model.add(Dense(10, kernel_initializer='he_uniform', activation='relu', input_dim=2))
model.add(Dropout(0.5))
model.add(Dense(8, kernel_initializer='he_uniform', activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(3, kernel_initializer='he_uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(1, kernel_initializer='glorot_uniform', activation='sigmoid'))

model.compile(optimizer='Adamax', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(x_new, y, validation_split=0.33, epochs=100)
```
